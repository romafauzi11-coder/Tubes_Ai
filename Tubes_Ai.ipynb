**Hitung nilai fuzzy servis**

```python
# Fungsi hitung nilai fuzzy servis
def hitung_servis(n):
    d = {"rendah": 0, "sedang": 0, "tinggi": 0}
    if n <= 45:
        d["rendah"] = 1
    elif 45 < n < 65:
        d["rendah"] = (65 - n) / (65 - 45)
        d["sedang"] = (n - 45) / (65 - 45)
    elif 65 <= n <= 75:
        d["sedang"] = 1
    elif 75 < n < 90:
        d["sedang"] = (90 - n) / (90 - 75)
        d["tinggi"] = (n - 75) / (90 - 75)
    else:
        d["tinggi"] = 1
    return d
```

**Hitung nilai fuzzy harga**

```python
# Fungsi hitung nilai fuzzy harga
def hitung_harga(n):
    d = {"murah": 0, "normal": 0, "mahal": 0}
    if n <= 3:
        d["murah"] = 1
    elif 3 < n < 5:
        d["murah"] = (5 - n) / (5 - 3)
        d["normal"] = (n - 3) / (5 - 3)
    elif 5 <= n <= 7:
        d["normal"] = 1
    elif 7 < n < 9:
        d["normal"] = (9 - n) / (9 - 7)
        d["mahal"] = (n - 7) / (9 - 7)
    else:
        d["mahal"] = 1
    return d
```

**Rules**

```python
# Proses inferensi rule base
def inferensi(f_servis, f_harga):
    t1 = min(f_servis["tinggi"], f_harga["murah"])
    t2 = min(f_servis["tinggi"], f_harga["normal"])
    t3 = min(f_servis["sedang"], f_harga["murah"])
    tinggi = max(t1, t2, t3)

    r1 = f_servis["rendah"]
    r2 = f_harga["mahal"]
    r3 = min(f_servis["sedang"], f_harga["normal"])
    rendah = max(r1, r2, r3)

    return rendah, tinggi
```
